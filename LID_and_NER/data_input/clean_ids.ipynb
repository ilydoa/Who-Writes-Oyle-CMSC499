Initial Dataset has duplicated doc_IDs. Notebook makes all doc_IDs unique. 

```python
import pandas as pd

def make_doc_ids_unique(df, id_column='doc_id'):
    """
    Groups contiguous blocks of the same doc_id and assigns a unique ID,
    ensuring merged files with overlapping IDs are split correctly.
    """
    # 1. Detect every time the doc_id changes from the previous row
    is_new_block = df[id_column] != df[id_column].shift()
    
    # 2. Create an overall sequential block number (1, 2, 3...)
    block_id = is_new_block.cumsum()
    
    # 3. Combine the original ID with the block number for uniqueness
    df[id_column] = df[id_column].astype(str) + '_' + block_id.astype(str)
    
    return df
```

```python
df = pd.read_csv("golden_annotated_dataset.csv")
df = make_doc_ids_unique(df, id_column='doc_id')
print(df)
```

```python
df.to_csv("cleaned_annotated_dataset.csv")
```
