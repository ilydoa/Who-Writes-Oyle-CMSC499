```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset name")
df.head()
```

```python
df['lid'].value_counts()
```

```python
df['lid'].value_counts().plot(kind='bar')
plt.title('Frequency of Each LID Label')
plt.xlabel('LID')
plt.ylabel('Count')
plt.show()
```

```python
df['ner'].value_counts()
```

```python
df['ner'].value_counts().plot(kind='bar')
plt.title('Frequency of Each NER Label')
plt.xlabel('NER')
plt.ylabel('Count')
plt.show()
```

```python
df['ner'].value_counts().drop("O").plot(kind='bar')
plt.title('Frequency of Each NER Label')
plt.xlabel('NER')
plt.ylabel('Count')
plt.show()
```
